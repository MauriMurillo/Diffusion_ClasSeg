```python
import matplotlib.pyplot as plt
import numpy as np
import glob
from PIL import Image
import cv2
import random
```

```python
r = np.random.randn(128, 128, 128, 3)
plt.hist(r[..., 0].flatten(), bins=50)
plt.hist(r[..., 1].flatten(), bins=50)
plt.hist(r[..., 2].flatten(), bins=50)
plt.show()
```

```python
images = glob.glob("/home/student/andrewheschl/Documents/Datasets/ClassificationDataset/preprocessed/Dataset_large_421/fold_0/train/imagesTr/*")
random.shuffle(images)

img = np.stack([cv2.resize(np.load(x)[:, 0:512, 0:512].transpose((1, 2, 0)), (128, 128), interpolation=cv2.INTER_CUBIC) for x in images[0:128]], axis=0)

plt.hist(img[:, :, :, 0].flatten(), bins=50, color='r')
plt.hist(img[:, :, :, 1].flatten(), bins=50, color='g')
plt.hist(img[:, :, :, 2].flatten(), bins=50, color='b')
plt.show()
```

```python
# normalized histogram
bins = 255
print(img.shape)
r_hist, bins_r = np.histogram(img[:, :, :, 0].flatten(), bins=bins)
r_hist=r_hist/img[...,0].size

g_hist, bins_g = np.histogram(img[:, :, :, 1].flatten(), bins=bins)
g_hist=g_hist/img[...,0].size

b_hist, bins_b = np.histogram(img[:, :, :, 2].flatten(), bins=bins)
b_hist=b_hist/img[...,0].size
plt.bar(bins_r[:-1], r_hist, color='r')
plt.bar(bins_g[:-1], g_hist, color='g')
plt.bar(bins_b[:-1], b_hist, color='b')
plt.show()
```
