```python
import torch
import numpy as np

path = "./embed.pt"
```

```python
embed = torch.load(path)
```

```python
print(embed.shape)
```

```python
embed = embed.cpu().numpy()
```

```python
import matplotlib.pyplot as plt

plt.scatter(embed[:, 0], embed[:, 1])
plt.show()

plt.scatter(embed[:, 0], embed[:, 2])
plt.show()
```

```python
# see covariance
cov = np.cov(embed.T)# heatmap
plt.imshow(cov, cmap='hot', interpolation='nearest')
# scale
plt.colorbar()
plt.show()
```

```python
og_im = torch.load("embed_sample.pt")
recons = torch.load("recon.pt")
og_im = og_im.cpu().numpy()
recons = recons.cpu().numpy()

print(recons.shape)
print(og_im.shape)

# show recons and original
# plt.imshow(og_im[0].transpose(1, 2, 0))
print(recons.max(), recons.min(), recons.mean(), recons.std())
_, ax = plt.subplots(2, 2)
ax[0, 0].imshow(og_im[0].transpose(1, 2, 0))
ax[0, 1].imshow(recons[0].transpose(1, 2, 0))
ax[1, 0].imshow(og_im[1].transpose(1, 2, 0))
ax[1, 1].imshow(recons[1].transpose(1, 2, 0))
plt.show()
```

```python
# plot all original
num = og_im.shape[0]
_, ax = plt.subplots(1, num, figsize=(20, 20))
for i in range(num):
    ax[i].imshow(og_im[i].transpose(1, 2, 0))
plt.show()
```
